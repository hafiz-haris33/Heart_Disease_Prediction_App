# Prepare the Data for Machine Learning Algorithms

## Create a Test Set
 First, we will create a test set to separate it from the rest of the data.
- The reason for creating a test set is to avoid bias towards the test set while preparing and training the model.
- By creating a test set at the beginning, we will hide it from ourselves and use it only for evaluating the model's performance.

We created a test set using stratified sampling.
- Stratified sampling ensures that the test set and training set have the same proportion of key features.
- This method helps maintain the same class distribution in both the training set and test set, which is essential for accurate model evaluation.

```python
from sklearn.model_selection import train_test_split

strat_train_set, strat_test_set = train_test_split(
    heart,
    test_size=0.2,
    random_state=42,
    stratify=heart['HeartDisease']
)
```

```python
strat_test_set["HeartDisease"].value_counts()/len(strat_test_set)
```

```python
heart["HeartDisease"].value_counts()/len(heart)
```

## Future Steps
- All future data preparation steps for machine learning algorithms will be performed using the training set.
- We will use the training set to prepare, transform, and feature-engineer the data.
- The test set will remain untouched until the final model evaluation stage.

### Splitting Data into Features and Labels
- We have split the training set into two parts:
    - heart: contains only the feature columns.
    - heart_labels: contains only the target variable or label column.

```python
heart = strat_train_set.drop('HeartDisease', axis=1)
heart_labels = strat_train_set['HeartDisease'].copy()
```

## Make pipelines

- We will create separate pipelines for numerical and categorical columns.
- For numerical columns, the pipeline will include:
    - SimpleImputer to handle missing values.
    - StandardScaler to scale the features.
- For categorical columns, the pipeline will include:
    - SimpleImputer to handle missing values.
    - A suitable encoding method (e.g., One-Hot Encoding) to transform categorical variables into numerical representations.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

num_pipeline = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)
```

## Merging Pipelines with ColumnTransformer
- We will import ColumnTransformer to combine the numerical and categorical pipelines.
- The ColumnTransformer will allow us to apply different preprocessing steps to different columns.
- We will create a preprocessing variable that combines both pipelines.
- Using the fit_transform method on the preprocessing variable, we can perform preprocessing on the entire dataset in a single step.

```python
from sklearn.compose import make_column_selector, make_column_transformer
import numpy as np

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),
)
```

```python
heart_prepared = preprocessing.fit_transform(heart)
```

# Fine-Tune Your Model

## Hyperparameter Tuning
- We will use Grid Search CV and Randomized Search CV to fine-tune the hyperparameters of both models.
- This will help us find the optimal combination of hyperparameters that results in the best performance for each model.
- After fine-tuning, we will compare the performance of both models again to select the final model.

### Logistic Regression

#### GridSearchCV

**Grid Search CV for Logistic Regression**
- Created a pipeline with preprocessing and logistic regression.
- Defined a parameter grid with different hyperparameters and their values.
- Applied Grid Search CV to the pipeline with the parameter grid.
- Fit the Grid Search CV to the training data.


```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# Pipeline: preprocessing + logistic regression
log_reg_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("log_reg", LogisticRegression())
])

param_grid = {
    'log_reg__C': [0.01, 0.1, 1, 10, 100],
    'log_reg__penalty': ['l1', 'l2'],
    'log_reg__solver': ['liblinear']  # 'liblinear' supports both l1 and l2
}


LR_grid_search = GridSearchCV(log_reg_pipeline, param_grid, cv=5, scoring='accuracy')

LR_grid_search.fit(heart, heart_labels)
```

```python
LR_grid_search.best_score_
```

```python
cv_res = pd.DataFrame(LR_grid_search.cv_results_)
cv_res.sort_values(by="mean_test_score", ascending=False, inplace=True)
cv_res.head()
```

- The best score obtained was approximately 0.85415, indicating the optimal hyperparameters for the logistic regression model.

#### Randomized Search

**Randomized Search CV for Logistic Regression**
- Imported RandomizedSearchCV and loguniform.
- Created a pipeline with preprocessing and logistic regression.
- Defined a randomized parameter distribution with different hyperparameters and their values.
- Applied Randomized Search CV to the pipeline.
- Fit the Randomized Search CV to the training data.


```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform

# Pipeline: preprocessing + logistic regression
log_reg_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("log_reg", LogisticRegression())
])

# Randomized parameter distribution
param_distributions = {
    'log_reg__C': loguniform(0.001, 100),       # Better for continuous range
    'log_reg__penalty': ['l1', 'l2'],
    'log_reg__solver': ['liblinear']  # because it supports both l1 and l2
}

# RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

LR_random_search = RandomizedSearchCV(
    log_reg_pipeline,
    param_distributions,
    n_iter=20,                # Try 20 random combinations
    cv=5,
    scoring='accuracy',
    random_state=42
)

# Fit the model
LR_random_search.fit(heart, heart_labels)
```

```python
LR_random_search.best_score_
```

```python
cv_res = pd.DataFrame(LR_random_search.cv_results_)
cv_res.sort_values(by="mean_test_score", ascending=False, inplace=True)
cv_res.head()
```

- The best score obtained was approximately 0.85416, indicating the optimal hyperparameters for the logistic regression model.

### RandomForestClassifier

#### Grid search

**Grid Search CV for Random Forest Classifier**
- Created a pipeline with preprocessing and Random Forest Classifier.
- Defined a hyperparameter grid with different values for the Random Forest Classifier's hyperparameters.
- Applied Grid Search CV to the pipeline with the hyperparameter grid.
- Fit the Grid Search CV to the training data.


```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Pipeline: preprocessing + Random Forest Classifier
rf_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("rf_clf", RandomForestClassifier(random_state=42))
])

# Hyperparameter grid for Random Forest Classifier
param_grid = {
    'rf_clf__n_estimators': [100, 200, 300],
    'rf_clf__max_depth': [None, 10, 20, 30],
    'rf_clf__max_features': ['sqrt', 'log2'],
    'rf_clf__min_samples_split': [2, 5, 10],
    'rf_clf__min_samples_leaf': [1, 2, 4]
}

# GridSearchCV
rf_grid_search = GridSearchCV(rf_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the model
rf_grid_search.fit(heart, heart_labels)
```

```python
rf_grid_search.best_score_
```

- The best score obtained was approximately 0.86372, indicating the optimal hyperparameters for the Random Forest Classifier model.

#### Randmized Search

**Randomized Search CV for Random Forest Classifier**
- Created a pipeline with preprocessing and Random Forest Classifier.
- Defined a hyperparameter distribution with different values for the Random Forest Classifier's hyperparameters.
- Applied Randomized Search CV to the pipeline.
- Fit the Randomized Search CV to the training data.


```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Pipeline: preprocessing + Random Forest Classifier
rf_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("rf_clf", RandomForestClassifier(random_state=42))
])

# Randomized hyperparameter distribution
param_dist = {
    'rf_clf__n_estimators': randint(100, 301),        # 100 to 300
    'rf_clf__max_depth': [None, 10, 20, 30],
    'rf_clf__max_features': ['sqrt', 'log2'],
    'rf_clf__min_samples_split': randint(2, 11),      # 2 to 10
    'rf_clf__min_samples_leaf': randint(1, 5)         # 1 to 4
}

# RandomizedSearchCV
rf_random_search = RandomizedSearchCV(
    rf_pipeline,
    param_distributions=param_dist,
    n_iter=20,               # Try 20 random combinations
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

# Fit the model
rf_random_search.fit(heart, heart_labels)

# Best result
print("Best Parameters:", rf_random_search.best_params_)
print("Best Score (Accuracy):", rf_random_search.best_score_)
```

- The best score obtained was approximately 0.86508, indicating the optimal hyperparameters for the Random Forest Classifier model.
